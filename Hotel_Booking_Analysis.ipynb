```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

# New Section

```python
# Load the dataset
data = pd.read_csv("hotel_bookings.csv")
```

```python
# Explore the first few rows of the dataset
print(data.head())
```

```python
print(data.describe())
```

```python
# Check for missing values
print(data.isnull().sum())
```

```python
# Data analysis and visualization

# Plot the count of bookings by hotel type
plt.figure(figsize=(8, 6))
sns.countplot(x='hotel', data=data)
plt.title('Count of Bookings by Hotel Type')
plt.xlabel('Hotel')
plt.ylabel('Count')
plt.show()
```

```python
# Plot the count of cancellations
plt.figure(figsize=(8, 6))
sns.countplot(x='is_canceled', data=data)
plt.title('Count of Cancellations')
plt.xlabel('Cancellation Status')
plt.ylabel('Count')
plt.show()
```

```python
# Plot the count of bookings by month
plt.figure(figsize=(12, 6))
sns.countplot(x='arrival_date_month', data=data, order=['January', 'February', 'March', 'April', 'May', 'June', 'July', 'August', 'September', 'October', 'November', 'December'])
plt.title('Count of Bookings by Month')
plt.xlabel('Month')
plt.ylabel('Count')
plt.xticks(rotation=45)
plt.show()
```

```python
# Plot the distribution of lead time
plt.figure(figsize=(8, 6))
sns.histplot(data['lead_time'], bins=50, kde=True)
plt.title('Distribution of Lead Time')
plt.xlabel('Lead Time')
plt.ylabel('Frequency')
plt.show()
```

```python
#Box plot of lead time by hotel type
plt.figure(figsize=(10, 6))
sns.boxplot(x='hotel', y='lead_time', data=data)
plt.title('Lead Time Distribution by Hotel Type')
plt.xlabel('Hotel')
plt.ylabel('Lead Time')
plt.show()
```

```python
plt.figure(figsize=(10, 6))
sns.countplot(x='market_segment', data=data)
plt.title('Count of Bookings by Market Segment')
plt.xlabel('Market Segment')
plt.ylabel('Count')
plt.xticks(rotation=45)
plt.show()
```

```python
plt.figure(figsize=(10, 6))
sns.lineplot(x='arrival_date_month', y='adr', data=data)
plt.title('Average Daily Rate (ADR) by Month')
plt.xlabel('Month')
plt.ylabel('ADR')
plt.xticks(rotation=45)
plt.show()
```

```python
plt.figure(figsize=(10, 6))
sns.histplot(data['total_of_special_requests'], bins=10, kde=True)
plt.title('Distribution of Total Special Requests')
plt.xlabel('Total Special Requests')
plt.ylabel('Frequency')
plt.show()
```
